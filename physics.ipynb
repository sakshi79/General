##### Model random motion of a grain of dust, according to this rule: I throw a dice. If I get between 1 and 3, the grain moves a step forward, otherwise it moves a step backward.

```python
from pylab import *
n = 1000
x = zeros(n, float)
for i in range(n-1):
    if (randint(6) + 1 <= 3):
        dx = 1
    else :
        dx = -1
    x[i+1] = x[i] + dx
    
```

```python
plot(x),
xlabel('i'), ylabel('x(i)')
show()
```

###### Now suppose the grain is enclosed in a box of width 10 units, i.e., The grain cannot move outside a region spanning from 5 to +5:

```python
n = 1000
x = zeros(n, float)
for i in range(n-1):
    if (randint(6) + 1 <= 3):
        dx = 1
    else :
        dx = -1
    x[i+1] = x[i] + dx
    if (x[i+1] > 5):
        x[i+1] = 5
    if (x[i+1] < -5):
        x[i+1] = -5

plot(x)
xlabel('i')
ylabel('x(i)')
show()
```

```python
#Aliter:
from pylab import *
n = 1000
x = zeros(n,float)
for i in range(1,n):
   if (randint(6)+1<=3):
      dx = -1
   else:
      dx = +1
   x[i] = x[i-1] + dx
   if (x[i]> 5):
      x[i] = 5
   if (x[i]<-5):
      x[i] = -5

plot(x),
xlabel('i'), ylabel('x(i)')
show()
```

```python
def normal(x, mu, sigma):
    P = (1/((2*pi*sigma**2)**0.5))*exp(-(x-mu)**2/(2*sigma**2))
    return P
```

```python
x = linspace(-5,5,1000)
normal_array = zeros(1000, float)
for i in range(1000):
    normal_array[i] = normal(x[i], 0, 1)
    
plot(x, normal_array)
show()
```

##### Logistic Map:
##### The iterative mapping x(i +1) = r x(i ) (1  x(i )) is called the logistic map.

```python
def logistic(x,r):
    value = r*x*(1-x)
    return value
```

```python
x_array = zeros(100, float)
x = 0.5
r = 1
for i in range(100):
    x_array[i] = x
    x = logistic(x,r) 
x_array[99]
```
